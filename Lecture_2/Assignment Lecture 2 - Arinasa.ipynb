## Task 1:
Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

Question 1: Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$

## Solution: Question 1

**Expected Value Calculation**

To compute the expected value $E[y]$, we utilize the joint probability distribution formula:

\[$ E[y] = \sum_{i=1}^{k} y_i \cdot P(y_i) $\]

Substituting the given joint probability distribution $P(y)$, we get:

\[$ E[y] = \sum_{i=1}^{k} y_i \cdot \left( \frac{n!}{y_1!y_2! \ldots y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot \ldots \cdot p_k^{y_k} \right) $\]

By expressing $y_i$ as $n$ minus the sum of all other $y_j$ terms, we can simplify:

\[$ E[y] = n \cdot \sum_{i=1}^{k} \left( \frac{(n-1)!}{y_1!y_2! \ldots y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot \ldots \cdot p_k^{y_k} \right) $\]

Since the term inside the summation represents the joint probability distribution for $(n-1)$ trials, the summation evaluates to 1:

\[$ E[y] = n \cdot 1 = n $\]

Thus, $E[y] = n \cdot p_i$, where $p_i$ is the probability of an outcome falling into cell $i$. 

Therefore, the expected value of the random variable $y$ is indeed $n \cdot p_i$.

**Probability Distribution Derivation**

The probability distribution of a binomial experiment is given by the binomial probability formula:

\[$ P(Y = k) = \binom{n}{k} \cdot p^{k} \cdot (1 - p)^{n-k} $\]

where:
- $k$ is the number of successes (the value that $Y$ takes)
- $\binom{n}{k}$ represents the number of ways to choose $k$ successes out of $n$ trials

## Solution: Question 2

**Expectation Calculation**

Using the probability distribution:

\[$ P(Y = k) = \binom{n}{k} \cdot p^{k} \cdot (1 - p)^{n-k} $\]

The expectation $E[Y]$ is calculated as:

\[$ E[Y] = \sum_{k=0}^{n} y \cdot P(Y = y) $\]

Substituting the PMF of the binomial distribution:

\[$ E[Y] = \sum_{k=0}^{n} k \cdot \binom{n}{k} \cdot p^{k} \cdot (1 - p)^{n-k} $\]

We can use the identity $k \cdot \binom{n}{k} = n \cdot \binom{n-1}{k-1}$ to simplify this expression. This identity states that choosing $k$
items out of $n$ and then picking one out of these $k$ as distinct can be equated to picking one out of $n$ as distinct and then choosing $k-1$
from the remaining $n-1$.

So, the expression becomes:

$E[Y] = \sum_{k=0}^{n} n \cdot \binom{n-1}{k-1} \cdot p^{k} \cdot (1-p)^{n-k}$

Now, notice that this expression resembles the binomial expansion of $(p + (1-p))^{n}$, which is $1^{n} = 1$.
Therefore, the sum simplifies to $n \cdot 1 = n$.

Thus,

$E[Y] = n \cdot p$<br><br>

## Solution: Question 3

**Probability of Recovery**

Given that 30% of all persons afflicted by a certain illness recover, let's analyze the scenario where ten people with the illness were selected at random and received medication, out of which nine recovered shortly thereafter. 

We'll use the binomial distribution formula to compute the probability of observing at least nine recoveries, assuming the medication is worthless.

### Probability Calculation

- **Binomial Distribution Formula**:

\[ P(Y = k) = \binom{n}{k} \cdot p^{k} \cdot (1 - p)^{n-k} \]

where:
- \( n \) is the number of trials (10 in this case)
- \( k \) is the number of successes (either 9 or 10)
- \( p \) is the probability of success (0.30 in this case)

### Calculations

- For \( k = 9 \):
  - \( P(Y = 9) = \binom{10}{9} \cdot (0.30)^{9} \cdot (0.70)^{1} \)
  - \( \approx 0.006006 \)

- For \( k = 10 \):
  - \( P(Y = 10) = \binom{10}{10} \cdot (0.30)^{10} \cdot (0.70)^{0} \)
  - \( \approx 0.000005904 \)

### Total Probability

Summing these probabilities:

\[ P(Y \geq 9) = P(Y = 9) + P(Y = 10) \]
\[ \approx 0.006006 + 0.000005904 \]
\[ \approx 0.006011904 \]

Therefore, the probability that at least nine out of ten people receiving the medication will recover, assuming the medication is worthless, is approximately \( 0.006011904 \).


## Solution: Question 4

**Probability Calculation**

To find the probability of observing at least one defective fuse in a sample of 5 fuses, we'll use the binomial distribution formula. In this case, the sample size is $n = 5$ and the proportion of defective fuses in the lot is $p = 0.05$. We want to calculate the probability of observing at least one defective fuse, which is the complement of the probability of observing no defective fuses.

First, we'll calculate the probability of observing no defective fuses, which is represented by $P(X = 0)$. Using the binomial distribution formula:

\[$ P(X = 0) = \binom{5}{0} \cdot (0.05)^0 \cdot (1 - 0.05)^{5} $\]
\[$ = \binom{5}{0} \cdot 1 \cdot (0.95)^5 $\]
\[$ = 1 \cdot (0.95)^5 $\]
\[$ \approx 0.77378 $\]

Now, to find the probability of observing at least one defective fuse, we subtract the probability of observing none from 1:

\[$ P(\text{at least one defective}) = 1 - P(\text{no defective}) $\]
\[$ = 1 - 0.77378 $\]
\[$ \approx 0.22622 $\]

Therefore, the probability of observing at least one defective fuse in a sample of 5 fuses is approximately 0.22622 or 22.622%.
