```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the number of nodes and edges
n = 10  # Number of nodes
m = 8   # Number of edges
seed = 42  # Seed for reproducibility

# Generate a random graph with n nodes and m edges
G = nx.gnm_random_graph(n, m, seed=seed)

# Add self-loops to the remaining nodes
edgelist = [(0, 0), (9, 9)]
G.add_edges_from(edgelist)

# Compute the layout for the nodes
pos = nx.spring_layout(G, seed=seed)

# Draw the graph with a background color gradient
plt.figure(figsize=(8, 6))
plt.subplot(111, facecolor='black')  # Background color
nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', edge_color='orange')
plt.title('Random Graph with Different Parameters')
plt.show()

# Get the edge-endpoint function
edge_endpoint_function = {f"e_{i}": set(edge) for i, edge in enumerate(G.edges())}

# Print the edge-endpoint function
print("Edge-Endpoint Function:")
print("Edge \t Endpoints")
for edge, endpoints in edge_endpoint_function.items():
    print(f"{edge} \t {endpoints}")
```

## Complete Graphs($K_n$)
- Loop Free
- Undirected
- For all vertices that are not equal to each other there is an edge between them <br>
  $\forall \, a,b \in V \, and \, a \neq b, \{a,b\} \in E$ <br>

  <b>Example:</b>

  $K_1 , K_2, K_3$

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_edges_from([('A','B'),('A','C'),('B','C'),('X','Y')])  
G.add_node ('T')

pos = nx.planar_layout(G)

nx.draw_networkx_nodes(G, pos, node_size=700)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='gold')
nx.draw_networkx_labels(G, pos)
plt.show()
```

### Bipartite Graphs

- $G$ is bipartite if we can split $V$ into components $V_a$ and $V_b$ where each $e \in E$ is of the form <br>
$\{v \in V_a, v \in V_b\}$ <br>
- Denoted as $K_{m,\,n}$ <br>

<b>Example: </b><br>

$K_{2,\,3}$

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_nodes_from([1,2], bipartite=0) # Add the node attribute "bipartite"
G.add_nodes_from([3,4,5], bipartite=1)
G.add_edges_from([(1,3), (1,4), (1,5)
                ,(2,3), (2,4), (2,5)])

top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top)
nx.draw(G, pos=pos, with_labels=True, edge_color = 'gold', node_color=['deepskyblue','pink','magenta','cyan','gold'])
plt.show()
```

## Subgraphs

Graph $H$ is said to be a subgraph of Graph $G$ <br>
$IF:$ <br>
1. Vertices of $H$ is also a vertex of $G$ <br>
    $(V_H \subseteq V_G)$

2. Edges of $H$ is also an edges of $G$ <br>
    $(E_H \subseteq E_G)$

3. Edges of $H$ are <b><i>incident</i></b> to vertices of $G$ <br>

<b>Example: </b>


```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
H = nx.Graph()
G.add_edges_from([('A','B'),('A','D'),('D','C'),('B','C'),('A','C'),('B','D')])
H.add_edges_from([('A','B'),('B','C'),('C','A')])

pos = {'A': (0, 0), 'B': (-1, 1), 'C': (0, 2), 'D': (1, 1)}

node_colors_G = ['lightblue', 'lightskyblue', 'lightcyan', 'lightgreen']
node_colors_H = ['lightcoral', 'lavender', 'lightpink']

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
nx.draw_networkx_nodes(G, pos, node_size=700, node_color=node_colors_G)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='gold')
nx.draw_networkx_labels(G, pos)
plt.title('Graph G')
plt.axis('off')

plt.subplot(1, 2, 2)
nx.draw_networkx_nodes(H, pos, node_size=700, node_color=node_colors_H)
nx.draw_networkx_edges(H, pos, edgelist=H.edges(), edge_color='gold')
nx.draw_networkx_labels(H, pos)
plt.title('Graph H')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## The Handshake Theorem

If $G$ is any graph, then the sum of the degrees of all the vertices of $G$ equals twice the number of edges of $G$. Specifically, if the vertices of $G$ are $v_1,v_2,...,v_n$, where $n$ is a nonnegative integer.<br><br>
The total degree of $G = deg(v_1) + deg(v_2 ) + · · · + deg(v_n )$ = $2 \, \cdotp \textbf{(the number of edges of G)}$.<br>

<b>Example:</b><br>

G is a simple graph with 24 edges and a degree of 4 for each vertex. Determine the total number of vertices.

<b>Solution:</b><br>

$\textbf{The number of edges} = 30$

$\textbf{Each vertex's degree} = 3$

$\textbf{Let n be the total number of vertices in the graph.}$

$\textbf{Sum of the degree of all vertices = 2 x Number of edges}$ 

$n \cdot 3 = 2 \cdot 30 \, \textbf{by substituting the variables.}$

$3n = 60$

$n = \frac{60}{3}$

$\textbf{Therefore, the total number of vertices in the graph is 20}$

## Euler's Circuit

- Traverses every edge once
- Every vertex has even degree

<b>Example: </b>

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a new graph
G = nx.Graph()

# Add nodes
for node in ['A', 'B', 'C', 'D']:
    G.add_node(node)

# Add edges between nodes to form an Eulerian circuit
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'A')]

# Add edges to the graph
G.add_edges_from(edges)

# Draw the network
plt.figure(figsize=(8, 6))
node_colors = ['lightblue' if node != 'B' else 'lightgreen' for node in G.nodes()]
nx.draw(G, with_labels=True, node_color=node_colors, edge_color='gold', font_weight='bold')
plt.show()
```

<b>Solution: </b><br>
$\textbf{A, B, C, D, A}$ <br>

## Hamiltonian Circuits

is a sequence of adjacent vertices and distinct edges in which every vertex of G appears exactly once, except for the first and the last, which are the same.

<b>Example:</b>

If a graph $G$ has a Hamiltonian circuit, then $G$ has a subgraph $H$ with the following properties:
1. $H$ contains every vertex of $G$.
2. $H$ is connected.
3. $H$ has the same number of edges as vertices.
4. Every vertex of $H$ has degree 2.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a new graph
G = nx.Graph()

# Add nodes
G.add_nodes_from("ABCDE")

# Add edges to form a Hamiltonian circuit
G.add_edges_from([("A", "B"), ("B", "C"), ("C", "D"), ("D", "E"), ("E", "A")])

# Define positions of nodes
pos = {
    "A": (0, 2),
    "B": (2, 2),
    "C": (2, 0),
    "D": (0, 0),
    "E": (1, 1)
}

# Draw the network
plt.figure(figsize=(8, 6))
nx.draw(G, pos=pos, with_labels=True, node_color="lightblue", node_size=2000, font_size=20, width=2, edge_color='gold')
nx.draw_networkx_labels(G, pos=pos, font_weight='bold', font_color='black', font_size=20)
plt.margins(0.2)
plt.show()
```

$\textbf{The Travelling Salesman Problem}$
---

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a new graph
G = nx.Graph()

# Add nodes
G.add_nodes_from(["A", "B", "C", "D", "E"])

# Add edges with distances
edgelist = [("A", "B", {'weight': 20}), ("A", "C", {'weight': 30}), ("A", "D", {'weight': 25}),
            ("B", "C", {'weight': 35}), ("B", "D", {'weight': 40}), ("B", "E", {'weight': 45}),
            ("C", "D", {'weight': 30}), ("C", "E", {'weight': 25}), ("D", "E", {'weight': 35})]
G.add_edges_from(edgelist)

# Position of nodes
pos = nx.spring_layout(G, seed=42)

# Draw the graph with edge labels
plt.figure(figsize=(8, 6))
nx.draw(G, pos=pos, with_labels=True, node_color="pink", node_size=2000, alpha=0.9, width=2, edge_color='gold')
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={('A', 'B'): '20', 
                 ('A', 'C'): '30', 
                 ('A', 'D'): '25',
                 ('B', 'C'): '35',
                 ('B', 'D'): '40',
                 ('B', 'E'): '45',
                 ('C', 'D'): '30',
                 ('C', 'E'): '25',
                 ('D', 'E'): '35'},
    font_color='red'
)

# Route examples
routes = {
    "ABCDE": 20 + 35 + 30 + 35,
    "ABDCE": 20 + 40 + 30 + 25,
    "ACBDE": 30 + 35 + 25 + 35,
    "ADEBC": 25 + 40 + 35 + 25,
    "AEBCD": 45 + 30 + 25 + 35
}

# Best route
best_route = min(routes, key=routes.get)

# Highlight best route
best_route_edges = [(best_route[i], best_route[i+1]) for i in range(len(best_route)-1)]
nx.draw_networkx_edges(G, pos, edgelist=best_route_edges, edge_color='blue', width=3)

# Print routes and best route
print("Route examples:")
for route, distance in routes.items():
    print(f"{route}: {distance}")
print("\nBest route is:", best_route)

plt.title("Traveling Salesman Problem")
plt.show()
```
